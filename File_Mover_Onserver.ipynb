```python
#script for moving files on the server and giving the right premisions 

# Import needed libraries
import os, json, csv, glob
from subprocess import call

# Path to the directory Clean
cleanDir = '/datb/aphasia/languagedata/uva/transform/clean/'
            

# This method gets all txt files
def getAllfiles(directory):
    return glob.glob(directory+'*.txt')

def moveFile(filepath, toFolder):
    moveCom = 'mv '
    call(moveCom + filepath + ' ' + toFolder, shell=True)
    
thres = 'a b c'

for i, file in enumerate(getAllfiles(cleanDir)):
    with open(file, 'r') as readfile:
        line = readfile.readline()
        line = readfile.readline()
        if thres in line.lower():
            moveFile(file, '/datb/aphasia/languagedata/uva/transform/excluded')
            print('Finished:{}'.format(file))
            
    
```

```python
#in dit nodebook worden de lostaande worden weg gehaald.
#ook wordt de text veder gecleand door de willekeurige 1 die soms voorkomen weg te halen.

import glob

folderOrig ="/datb/aphasia/languagedata/uva/transform/clean/*.txt"

import os


def remove_solowords_lines(filename):
    files = glob.glob(filename)
    for file in files:
    
        
        with open(file, "r") as filehandle:
            lines = filehandle.readlines()
    
        for idx, elem in enumerate(lines):
            if idx == 0:
                lines.insert(0,"1\r\n")
            if " " not in elem:
                if idx != 0:
                    lines[idx-1] = lines[idx-1].replace("\n","")
                    lines[idx-1] += ' ' + lines[idx]
                    print("lines[idx] " + lines[idx])
                    print("lines[idx-1] " + lines[idx-1])
                    del(lines[idx])

                
        with open(file, 'w') as filehandle:
            filehandle.writelines(lines) 


remove_solowords_lines(folderOrig)
```
